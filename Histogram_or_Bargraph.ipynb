```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['0-10', '11-20', '21-30', '31-40', '41-50']
male_counts = [10, 15, 20, 25, 10]
female_counts = [12, 18, 22, 20, 10]
non_binary_counts = [2, 4, 3, 1, 0]

x = np.arange(len(categories))  # the label locations
width = 0.2  # the width of the bars

# Creating the grouped bar chart
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width, male_counts, width, label='Male', color='blue')
rects2 = ax.bar(x, female_counts, width, label='Female', color='pink')
rects3 = ax.bar(x + width, non_binary_counts, width, label='Non-binary', color='purple')

# Adding labels, title and custom x-axis tick labels
ax.set_xlabel('Age Range')
ax.set_ylabel('Count')
ax.set_title('Age Distribution by Gender')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Displaying the chart
plt.show()
```
